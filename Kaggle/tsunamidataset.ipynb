```python
#replace zeros  with NAN
col_with_zeros=['nst','dmin','gap']
for col in col_with_zeros:
    df[col]=df[col].replace(0,np.nan)
    #impute Nan with median
    if df[col].isnull().sum()>0:
        median_val=df[col].median()
        df[col].fillna(median_val,inplace=True)
        print(f"-Imputed {col} NaN values with median:{median_val:.2f}")
print("- Missing values after imputation")
print(df.isnull().sum())
```

```python
#check for duplicates 
duplicate_count=df.duplicated().sum()
print(f"-Found {duplicate_count} duplicates rows")
if duplicate_count>0:
    df.drop_duplicates(inplace=True)
    print(f"-Dropped duplicates row {df.shape}")
```

```python
#handling outliers
print("- checking 'magnitude','depth','cdi','mmi' for extreme values using IQR method")
outliers_cols=['magnitude','depth','cdi','mmi']
for col in outliers_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
print(f"- {col}:{len(outliers)} outliers dtected (values <{lower_bound:.2f} or >{upper_bound:.2f}")
#cap outliers
df[col]=df[col].clip(lower=lower_bound,upper=upper_bound)
print(f"- Capped {col} outliers to bound [{lower_bound:.2f},{upper_bound:.2f}]")
    
```

.clip(lower=lower_bound, upper=upper_bound)

Any value less than lower_bound is replaced with lower_bound.

Any value greater than upper_bound is replaced with upper_bound.

Values within the range remain unchanged.

```python
#validate data types
#ensure that tsunami in binaryb and numerical col in int/float
df['tsunami']=df['tsunami'].astype(int)
numerical_cols=['magnitude','cdi','mmi','sig','nst','dmin','gap','depth','longitude','latitude']
for col in numerical_cols:
    df[col]=df[col].astype(float)
    print(f"-converted {col} in float")
```

```python
#validate ranges /normalization
#magnitude generally in range 6.5 to 9.5 for significae quakes
df=df[(df['magnitude']>=6.5 )& (df['magnitude']<=10.0)]
#Depth ,non-negative resonable max( e.g 700)
df=df[(df['depth']>=0)&(df['depth']<=700)]
print(f"filtered depoth {df.shape}")
```

```python
#cdi and mmi: 0-10 and 1-12 respectively
df=df[(df['cdi']>=0)&(df['cdi']<=10) & (df['mmi']>=1)&(df['mmi']<=12)]
print(f"-filtered cdi New shape{df.shape}")
```

```python
#tsunami ensure in binary form (0,1)
df=df[df['tsunami'].isin([0,1])]
print(f"-Ensured tsunami is bionary(0,1) new shape {df.shape}")
```

```python
#feature enginnering 
print("- Creating a new feature 'is_coastal' based on proximity to ocean")
#simple heuristic :earthquake within 200km of coastalines can cause tsunami
df['is_coastal']=((df['latitude'].abs()<60)&(df['longitude'].abs()<180).astype(int))
print("- Added 'is_coastal' feature (1=coastal,0=inland,based on latitude and longitude")
```

```python
# final check 
print("-Final shape of cleaned dataset",df.shape)
print("-final missing values:")
print(df.isnull().sum())
print("-fnal data types:")
print(df.dtypes)
print("\n saving cleaned dataset to 'earthquake_data_cleaned.csv'.")
df.to_csv('earthquake_data_cleaned.csv',index=False)
print("clened dataset has save successfully !")
```
